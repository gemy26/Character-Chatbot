````python
def save_data_o_file(character_name, character_info):
    filename = f"{character_name}.txt"
    with open(filename, "w", newline="", encoding="utf-8") as file:
        writer = csv.writer(file)
        writer.writerows(character_info)
````

````python
def get_character_info(character_name):
    character_name.replace(" ", "_")
    character_wiki_url = f'https://friends.fandom.com/wiki/{character_name}'
    site = requests.get(character_wiki_url)
    soup = BeautifulSoup(site.content, 'lxml')

    headers = soup.find_all(["h2", "h3"])  # Get all H2 and H3 headers
    result = []

    for header in headers:
        key = header.get_text(strip=True).replace("[edit]", "").strip()[:-2]  # Clean header text
        for sibling in header.find_next_siblings():
            if sibling.name in ["h2", "h3"]:
                break  # Stop when reaching a new header
            if sibling.name == "p":
                result.append([key, sibling.get_text(strip=True)])  # Store each paragraph separately

    return result
````

````python
for character in characters:
    characters_info = get_character_info(character)
    save_data_o_file(character, characters_info)
    print(len(characters_info))
````

````python
import json
import os
from openai import OpenAI
from google.colab import files

# Set API key
os.environ["OPENROUTER_API_KEY"] = "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"
api_key = os.getenv("OPENROUTER_API_KEY")
if not api_key:
    raise ValueError("API key is missing. Set it using os.environ['OPENROUTER_API_KEY']")

# Initialize OpenAI Client
client = OpenAI(
    base_url="https://openrouter.ai/api/v1",
    api_key=api_key,
)

def process_biography_file(character_name):
    file_path = f"/content/{character_name}.txt"
    output_path = f"/content/{character_name}_2.json"

    if not os.path.exists(file_path):
        print(f"ðŸš¨ Error: File {file_path} not found.")
        return

    try:
        with open(file_path, "r", encoding="utf-8") as f:
            biography = f.read().strip()

        if not biography:
            print(f"âš ï¸ Error: Biography file {file_path} is empty.")
            return

        prompt = f"""
        Extract structured data from the following character biography. Follow the instructions carefully:

        Biography:
        {biography}

        Instructions:
        - Extract the following fields from the biography:
          - name: The full name of the character.
          - birth_year: The year the character was born.
          - death_year: The year the character died (if applicable).
          - profession: The primary profession of the character.
          - traits: A list of personality traits of the character.
          - all_events: A list of **at least 5 events** related to the character, each with the following fields:
            - year: The year the event occurred.
            - event: The name of the event.
            - description: A brief description of the event.
            - quotes: A list of quotes related to the event.
            - characters: A list of other characters involved in the event.
            - location: The location where the event occurred.
            - event_type: The type of event (e.g., personal, professional, historical).
            - event_category: The category of the event (e.g., birth, marriage, battle, achievement).
            - event_sub_category: A sub-category of the event (if applicable).
          - major_events: A list of **at most 5 major events** in the character's life, each with the following fields:
            - year: The year the event occurred.
            - event: The name of the event.
          - personal_qualities: A list of personal qualities of the character.
          - relationships: A dictionary of relationships, including:
            - father: The character's father.
            - teacher: The character's teacher.
            - boyfriend: The character's boyfriend.
            - girlfriend: The character's girlfriend.
            - sibling: The character's sibling.
            - spouse: The character's spouse.
            - parent: The character's parent.
            - child: The character's child.
            - friend: The character's friend.
          - values: A list of values the character holds.
          - quotes: A list of notable quotes by the character.

        - Return ONLY valid JSON. No explanations, markdown, or additional text.
        - Ensure all lists have valid commas and brackets.
        - Do NOT include '```json' or any markdown formatting.
        - Return data strictly in the following format:

        {{
          "name": "",
          "birth_year": "",
          "death_year": "",
          "profession": "",
          "traits": [],
          "all_events": [
            {{
              "year": "",
              "event": "",
              "description": "",
              "quotes": [],
              "characters": [],
              "location": "",
              "event_type": "",
              "event_category": "",
              "event_sub_category": ""
            }}
          ],
          "major_events": [
            {{
              "year": "",
              "event": ""
            }}
          ],
          "personal_qualities": [],
          "relationships": {{
            "father": "",
            "teacher": "",
            "boyfriend": "",
            "girlfriend": "",
            "sibling": "",
            "spouse": "",
            "parent": "",
            "child": "",
            "friend": ""
          }},
          "values": [],
          "quotes": []
        }}
        """

        response = client.chat.completions.create(
            model="meta-llama/llama-3.3-70b-instruct:free",
            messages=[{"role": "user", "content": prompt}],
            temperature=0.3
        )

        # Get raw response
        raw_result = response.choices[0].message.content.strip()


        print("RAW API RESPONSE:\n", raw_result)

        # Handle markdown formatting (e.g., ```json)
        if raw_result.startswith("```json") and raw_result.endswith("```"):
            raw_result = raw_result[7:-3].strip()  # Remove ```json and ```

        # Attempt JSON extraction
        try:
            parsed_data = json.loads(raw_result)
        except json.JSONDecodeError as e:
            print(f"JSON parsing error: {e}\n")
            print("Problematic JSON content:\n", raw_result)
            return

        # Save structured data
        with open(output_path, "w", encoding="utf-8") as f:
            json.dump(parsed_data, f, indent=2)

        print(f"Successfully processed data. Saved to {output_path}")
        files.download(output_path)

    except Exception as e:
        print(f"An error occurred: {e}")
````
