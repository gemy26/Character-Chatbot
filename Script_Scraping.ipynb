**This Script to scrap script dialouge of friends tv-show**



```python
from bs4 import BeautifulSoup
import requests
import os
```

```python
url = 'https://edersoncorbari.github.io/friends/'
site = requests.get(url)
soup = BeautifulSoup(site.content, 'lxml')

get_and_save_content(soup)
```

```python
def get_and_save_content(soup):
    seasons = soup.find_all("h2")

    for season in seasons:
        season_title = season.get_text(strip=True)

        episode_list = season.find_next_sibling('ul')
        if not episode_list:
            continue
        for episode in episode_list.find_all('li'):
            episode_link = episode.find('a')

            if episode_link:
                episode_title = episode_link.get_text(strip=True)
                episode_url = episode_link['href']
                dialogue = get_dialogue(episode_url)
                save_to_file(season_title, episode_title, dialogue)


```

```python
def get_dialogue(url):
  page = requests.get('https://edersoncorbari.github.io' + url)
  soup = BeautifulSoup(page.content, 'lxml')
  dialogue = ""
  for line in soup.find_all('p'):
    character = line.find('b')
    if(character):
      character = character.text.strip()
    # print(line.text, character)
      dialogue += (character + line.text.strip())
    else:
      dialogue += line.text.strip()
    dialogue += "\n"
  return dialogue
```

```python
dialogue = get_dialogue('friends-scripts/season/0101.html')
print(dialogue)
```

```python
def save_to_file(season, episode_title, dialogue):
  folder_path = os.path.join("/Friends-Scripts", season)
  if not os.path.exists(folder_path):
    os.makedirs(folder_path)

  episode_title = episode_title.replace("/", "-")
  file_path = os.path.join(folder_path, f"{episode_title}.txt")
  print(season, episode_title)
  print(file_path)
  with open(file_path, "w", encoding="utf-8") as file:
    file.write(dialogue)
    print(f"{episode_title} saved in {file_path}")
```
